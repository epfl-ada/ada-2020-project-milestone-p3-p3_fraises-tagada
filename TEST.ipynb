```python
import os
import pandas as pd
import numpy as np
import warnings
import datetime as dt
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
PATH = 'Logit_Files/'
csv_names = os.listdir(PATH)
```

```python
fig , axis = plt.subplots(2,5, figsize = (20,10))

for name, ax in zip(csv_names,axis.flatten()):
    df = pd.read_csv(PATH+name)
    df.date = pd.to_datetime(df.date)
    Monthly = df.groupby(['subject_race', pd.Grouper(key = 'date', freq = '1M')]).mean()
    del df
    sns.lineplot(data = Monthly, hue = 'subject_race', x = 'date', y = 'search_rate', ax = ax)
    ax.set_title(name[:2])
```
